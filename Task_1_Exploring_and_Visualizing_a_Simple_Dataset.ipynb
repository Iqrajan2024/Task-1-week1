## *Importing the required libraries*

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## *Loading the dataset*

```python
# Load the Iris dataset available through seaborn library
df = sns.load_dataset('iris')
```

## *Exploratory Data Analysis (EDA)*

## *Scatterplots*

```python
# Plotting scatterplots to analyze the relationships between the variables/features using pairplot
sns.set_style('whitegrid')
sns.pairplot(df, hue='species')
plt.suptitle("Scatterplots of Iris features", y=1.02, fontsize=16, weight='bold', style='italic')
plt.show()
```

## *Histograms*

```python
# Drop 'species' to just get the numerical features/variables
features = df.drop(columns=['species'])

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Plot histogram for each feature by species
for i, col in enumerate(features):
    sns.histplot(data=df, x=col, hue='species', kde=True, bins=20, ax=axes[i], edgecolor='black')
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()
plt.suptitle("Distribution of Iris features", y=1.02, fontsize=16,  weight='bold', style='italic')
plt.show()
```

## *Box plots*

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Plot box plot for each feature by species
for i, col in enumerate(features):
    sns.boxplot(y=col, hue='species', data=df, ax=axes[i])
    axes[i].set_title(f"{col} ")

# Adjust layout
plt.tight_layout()
plt.suptitle("Boxplots of Iris Features", y=1.02, fontsize=16,  weight='bold', style='italic')
plt.show()
```
